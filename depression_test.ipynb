```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense
import math
```

```python
df = pd.read_csv('dep_clean.csv')
```

```python
plt.subplots(figsize=(9, 9))
sns.heatmap(df.corr(), annot=True)
```

```python
x = df.drop("dep", axis=1)
y = df.dep
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
model = Sequential()
model.add(Dense(64, input_dim=11, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(3, activation='sigmoid'))
model.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=200, batch_size=10)
```

```python
man = np.array([[0,3,3,25,0,0,2,2,1,1,2]])


out=model.predict(man)
h=np.array(out[0])
bars = ('low' , 'medium' , 'high')

plt.bar(bars , h)

plt.show()
```
